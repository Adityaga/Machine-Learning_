```python
#importing all the required libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df_x=pd.read_csv("../Downloads/Linear_X_Train.csv")   # collecting values of x and y into dataframe
df_y=pd.read_csv("../Downloads/Linear_Y_Train.csv")
print(df_x)
x=df_x.values
y=df_y.values
x=x.reshape((-1,1))
y=y.reshape((-1,1))
plt.scatter(x,y,label='Before scalling')
X=(x-x.mean())/x.std()  # feature scalling
Y=y
plt.scatter(X,Y,label='after scalling')
plt.legend()   # for printing labels 
plt.show()
```

```python
model=LinearRegression()    # clling model function of sklearn library
model.fit(X,Y)
model.coef_       #estimated coff of hypothesis/linear regression
   
```

```python
model.intercept_ # independent term of linear regression
```

```python
print(model.score(X,Y))
output = model.predict(X)   # for pridicted values of Y
plt.scatter(X,Y,label='data')
plt.plot(X,output,color='orange',label='prediction')
plt.legend()
plt.show()
print(output)
```

```python
df_test=pd.read_csv("../Downloads/Linear_X_Test.csv")  #importing testing data
test_values=df_test.values
test_val=(test_values-test_values.mean())/test_values.std()  #feature scalling
test_val=test_val.reshape((-1,1))
prediction=model.predict(test_val)  # for pridiction of values of Y
plt.plot(X,output,color='orange',label='learning')
plt.plot(test_val,prediction,color='blue',label='prediction')
plt.legend()
print(prediction)
```
